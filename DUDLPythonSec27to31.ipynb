```python
### Classes and Object Oriented Programming "OOP"
# Class: A blueprint for creating an object with set of attributes(like variables) and methods (like functions)
    # Instance: Classes may run multiple 'Instances'
    # PYTHON OBJECTS HIERARCHY:
      #1.CLASS: Software which encapsulates data and functions that operate on that data.
        #2.FUNCTION: Blocks of reusable code that perform a specific task.
          #3.DICTIONARIES:
              # Key:Value matrix
              # Data Structures here and below
            #4.LISTS:
                # Item and indexed data
              #5.TUPLES:
                #Item and data
### Example:
class model(object): # creates an object (type) object called 'model'
  #Constructor Method (foundational class invocation):
  def __init__(self,numlayers,numunits,name): # Methods & Functions
    self.layers = numlayers #Attribute/variable
    self.units = numunits #Attribute/variable
    self.name = name #Attribute/variable
    self.weights = 1 #This is what is often manipulated/tuned when the model "learns"

  def howManyUnits(self):
    totalUnits = self.layers * self.units
    print(f'There are {totalUnits} units in the model')

  def trainTheModel(self,x): # adds weight as it runs
    self.weights += x
    return self.weights
  def __str__(self):
    return f'This is a {self.name} architecture with {self.layers} layers and {self.units} units per layer'
mod1 = model(12,5,'CNN')
str(mod1)
# mod1.howManyUnits()

# mod1.trainTheModel(2.5)



```

```python
## EXCERCISE: Class

import random
import numpy as np


class model2(object): # creates an object (type) object called 'model'
  #Constructor Method (foundational class invocation):
  def __init__(self,numlayers,numunits,name):
    self.layers = numlayers # Attributes
    self.units = numunits
    self.name = name
    self.weights = np.random.randn(numlayers, numunits) # Read NOTES:
        # When to Use numpy.random.rand
            # Weight Initialization: Commonly used in machine learning to initialize weights in neural networks, ensuring they start from a uniform range.
            # Simulations: Useful for Monte Carlo simulations where uniform randomness is required.
            # Testing: Generating random test data for algorithms.
        # When to Use numpy.random.randn
            # Statistical Analysis: Ideal for generating data that follows a normal distribution, which is common in many statistical methods.
            # Machine Learning: Used in algorithms that assume normally distributed data, such as certain regression models.
            # Random Sampling: Useful for creating samples that need to reflect real-world phenomena, which often follow a normal distribution.

  def howManyUnits(self):
    totalUnits = self.layers * self.units
    print(f'There are {totalUnits} units in the model')

  def trainTheModel(self,x,y): # adds weight as it runs
    self.weights*x + y

  def __str__(self):
    return f'This is a {self.name} architecture with {self.layers} layers and {self.units} units per layer'

```

```python
### Fibonacci Series Calculator ###
## Define a Function that will output the Fibonacci sequence/series for input N

def fibseq(N):
    fn = [0,1]
    for i in range(N-2):
        fn.append( sum(fn[-2:]) ) # .append modifies a list and iteratively sums the last 2 numbers in the range
    print(fn)
fibseq(5)
```

```python
### Run three ways to initialize the variables. Test to see how computationally intensive each method of creating an array actually is by timing it using 'time.perf_counter':
import time
import numpy as np

# 1. Time test Numpy initialization (ndarray: using np.zeros)
    # SYNTAX: time.perf_counter()
start_time = time.perf_counter()
N = 100000
r = np.zeros(N)
for i in range(N):
    r[i] = i**2
#print(r)
stop_time = time.perf_counter()
print('Elapsed time for "np.zeros intialization":', 1000*(stop_time - start_time), 'ms')
type(r)
```

```python
import time
import numpy as np

# 2. Time test Numpy initialization (ndarray: using np.full)
    # SYNTAX: time.perf_counter()
start_time = time.perf_counter()
N = 100000
r = np.full(N,0)

for i in range(N):
    r[i] = i**2
stop_time = time.perf_counter()
print('Elapsed time for "np.full initialization":', 1000*(stop_time - start_time), 'ms')
type(r)
```

```python
### Mean and Variance ###

# import libraries
import numpy as np
import torch
import torch.nn as nn

# Create a list of values to compute with
x = [1,2,4,6,5,4,0]
n = len(x)

# Compute the mean
mean1 = np.mean(x)
mean2 = np.sum(x)/n

#Outputs
print(mean1)
print(mean2)
```

```python
### var1 and var 2 should give same variance value but the numpy version in var1 has "biased" variance
## Denominator N is biased (0 degrees of freedom)
## Denominator N-1 is unbiased (1 degree of freedom)
## Bias (degrees of freedom) is only a big problem in small data sets
var1 = np.var(x)
var2 = ( 1/(n-1)) * np.sum( (x-mean1)**2 )

print(var1)
print(var2)
```

```python
## Solving the bias problem by specifying the number of degrees of freedom for np.var


var3 = np.var(x,ddof=1)
print(var3)
print(var2)
```

```python
### Larger sample sizes are less prone to skewing the Variance with bias 1000 samples

N=1000
x1 = np.random.randint(0,high=20,size=N)

var01=np.var(x1,ddof=0) # np default, biased
var02=np.var(x1,ddof=1) # unbiased

print(var01)
print(var02)
```

```python
### The T-Test ####

## Helps to determine which Architecture and set of meta-parameters give the  most accurate results
## Helps determine to what extent samples drawn from one distribution or another are similar
## Provides a means to compare the results for different models

import scipy.stats as stats

import numpy as np
import matplotlib.pyplot as plt


```

```python
# Parameters

n1 = 30 # Number of samples in DataSet 1
n2 = 40 # Number of samples in DataSet 2
mu1 = 1 # Population mean in DataSet 1
mu2 = 2 # Population mean in DataSet 2

## Generate data #
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

## Set up the Plot #

plt.plot(np.zeros(n1),data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2),data2,'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1','Group 2'])
plt.show()

## but... are the differences below statisically significant?
```

```python
## Build and evaluate the t-test results ##

# Independent Samples (ttest_ind) Other ttest variants are available
t,p = stats.ttest_ind(data1,data2)

print(t)
print(p)

## data1 has a much larger Mean than data2:
# t: -3.0886743113566406
# p:  0.002910564077190452 are the differences below statisically significant? YES the "P" Value is <0.05!
## Change the "mu" values above to change the "P" value to see more or less
```

```python
### Derivative Product Rule and Chain rules ###

import numpy as np
import sympy as sym
from IPython.display import display

## Symbolic variables:
x = sym.symbols('x')

## Create the functions
fx = 2*x**2
gx = 4*x**3 - 3*x**4

## Compute the function Derivatives:
df = sym.diff(fx)
dg = sym.diff(gx)

## Product rule manual steps
manual = df*gx + fx*dg
thewrongway = df*dg

# via SymPy
viasympy = sym.diff( fx*gx )

## Print out
print("The Functions: ")
display(fx)
display(gx)
print(' ')

print("Their Derivatives: ")
display(df)
display(dg)
print(' ')

print("The Manual Product Rule: ")
display(manual)
print(' ')

print("Via SymPy Library: ")
display(viasympy)
print(' ')

print("The Wrong Way (violates the Product Rule): ")
display(thewrongway)
print(' ')
```

```python
## Function Definitions ##
# Define the x function
def fx(x):
    return 3*x**2 - 3*x + 4

# Derivative of x function
def deriv(x):
    return 6*x - 3
```

```python
## Plot the function  and its derivative:

# define the range of x
x = np.linspace(-2,2,2001)

# Plot the graphs:
plt.plot(x,fx(x), x,deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()

```

```python
## Random starting point for the function ##
localmin = np.random.choice(x,1) # choose 1 random choice value from vector x
print(localmin)
## Learning paraameters

learning_rate = .01
training_epochs = 100

## Run through the training

for i in range(training_epochs):
    grad = deriv(localmin) # take a derivative guess at the Local Minimum
    localmin = localmin - learning_rate*grad # then adjust the Local Minimum to be itself minus the learning rate * the guessed gradient value for 100 epochs
localmin
```

```python
## Plot the results ##

plt.plot(x,fx(x), x,deriv(x))
plt.plot(localmin,deriv(localmin), 'ro')
plt.plot(localmin,fx(localmin), 'ro')
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical Local Minimum: %s'%localmin[0])
plt.show()
```

```python
### Repeat but include every localmin value as the FOR Loop cycles through each training epoch ###

localmin = np.random.choice(x,1) # choose 1 random choice value from vector x
print(localmin)
## Learning parameters

learning_rate = .1
training_epochs = 7

## Run through the trainingm and store the results

modelparams = np.zeros((training_epochs,2))
for i in range(training_epochs):
    grad = deriv(localmin) # take a derivative guess at the Local Minimum
    localmin = localmin - learning_rate*grad # then adjust the Local Minimum to be itself minus the learning rate * the guessed gradient value for 100 epochs
    modelparams[i,0] = localmin[0]
    modelparams[i,1] = grad[0]

localmin
```

```python
## Plot the gradient over 100 itertions
fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
    ax[i].plot(modelparams[:,i],'o-')
    ax[i].set_xlabel('Iteration')
    ax[i].set_title(f'Final Estimated Minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local Minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

```python
### EXCERCISE: Poisson Distr. Counter ###

# 1: Specify a Target number of e^(-L) OR np.exp(-lambda) --> set = 10 initially
# 2: Initialize the 'counter' as '0', 'currval' as 1
# 3: multiply the currval by a random number U[0,1] np.randint
# 4: Repeat step 3 until currval < target
# 5: Return how many iterations it took through steps 3-4

import numpy as np

def poisson_ctr(lam):
# Intialize values
    counter,currval = 0,1
    target = np.exp(-lam)

# implement algorithm
    while currval > target:
        counter += 1
        currval *= np.random.rand() # generates random number from a uniform distribution from 0 up to 1 (instead of a normal (n) range centered on 0)
    return counter

# return the result


```

```python
## Modified to separate out the newly created matrix
import numpy as np

Y = np.zeros(X.shape)

X = np.array([ [1,2,3], [2,3,4], [3,4,5], [4,5,6] ])

w = np.array( [10,20,30] )

for i in range(Y.shape[0]):
    Y[i,:] = X[i,:] + [w]

Y

```

```python
### Broadcasting operations #256: ###
## Allows math operations to automatically proceed by allowing minimalist math syntax for vectors and matrices if the shapes are compatible

# Row Broadcasting:

Z = X+w # Using Addition

Z = X-w # Using Subtraction

Z = X*w # Using Multiplication

print(Z)
print()
print(Y)

```

```python
# Column Broadcasting:
import numpy as np
X = np.array([ [1,2,3], [2,3,4], [3,4,5], [4,5,6] ])

# V = np.array( [-1,0,1,0] ) # Row Vector

##  Convert the row vector to a column vector: ##

# V = np.array( [[-1],[0],[1],[0]] ) # Way 1: make each value into individual rows

V = np.array( [-1,0,1,0],ndmin=2 ).T # Way 2: Specify 2 dimensions and transpose (.T) the vector:
# X+V
print(X), print()
print(V)
X*V
```

```python
### Broadcasting Excercise ###

# 1: Create a vector (v) of integers 0:8 (0 to 8) x
# 2: Reshape (v) into a 3x3 matrix (m). use: np.reshape
# 3: Repeat (m) into a 9x3 matrix (C). use: np.tile
# 4: Broadcasting: multiply C with v to make (B)

import numpy as np
v = np.arange(9)
print(v)
print()
m = np.reshape(v,(3,3))
print(m)
print()
C = np.tile(m,(3,1))
print(C)
print()
# (v) must be converted to a column vector (V) so the shapes will be compatible for broadcasting
V = np.array( v,ndmin=2 ).T
B = C*V # Numpy broadcasting...
print(B)
print()
```

```python
### Functional error checking/handling #257 ###

# Create a function that will input 2 numbers and output their product

def funfun(x,y):
    # verify that inputs are :
    if not isinstance (x, (int,float)):
        raise Exception('Input "x" must be a number')
    if not isinstance (y, (int,float)):
        raise Exception('Input "y" must be an number')
    z = x * y
    return z

funfun("2",3)
```

```python
def funfun_simpler(x, y):
    try:
        # Convert both inputs to float (works for int, float, and numeric strings)
        x = float(x)
        y = float(y)
        z = x * y
        return z
    except (ValueError, TypeError) as e:
        raise Exception(f'Inputs must be numbers or numeric strings: {e}')

print(funfun_simpler("2", 3))
print(funfun_simpler("2.5", "4.0"))
print(funfun_simpler(1, 2.2))
```

```python
### Printing and string Interpolation ###

## Original method:

num = 11.754932
strg = 'Cellar Doors'
print('I would like to have ' + str(num) + ' ' + strg + '.')
```

```python
## Using % Variables

print('I would like to have %4.3f %s' %(num,strg))

# Different % Variables:
# %g = number 'as-is' (float or integer)
# %s = string
# %i = integer
## Float Options
### %f = float (greater number of decimal spaces)
    # %.{x}f = float (set the decimal spaces and round to the last place)
        # ' %.3f ' for 11.7 = 11.700 (float w/ 3 decimal places)
        # ' %.3f ' for 11.754932 = 11.755
        # Limiting to 1 ( %.1f ) or 2 ( %.2f ) decimal places allows easier graph/plotting
    # ' %.0f ' will round to the nearest full #/integer
        # ' %.0f '= 12 (from 11.7)
    # '%{x}.{y}f = sets {x} (total digit count)
        # ' %5.3f = for 11.754932 = 11.755
        # NOTE: ' %4.3f still = 11.755 (5 sig digits, when 4 is what the code shows)
```

```python
## F-String Interpolation ##
print(f'I would like to have {num} {strg}')

# {num:.2f} will do the same as the % variable option:
# print(f'I would like to have {num:.2f} {str}')
        # <I would like to have 11.75 Cellar Doors>
```

```python
### Plotting with Seaborn ###

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

n = 200
D = np.zeros((n,2)) # Defines a new dummy matrix of 200 (rows) x 2 (column) full of zeros
D[:,0] = np.linspace(0,10,n) + np.random.randn(n) # Fill the first column (column 0) with linearly spaced numbers from 0 to 10 in 'n' (200) steps + noise
D[:,1] = D[:,0]**2 + np.random.randn(n)*10 # Fill the second column (column 1) with linearly spaced numbers from 0 to 10 in 'n' (200) steps and add noise (*10)

sns.jointplot(x=D[:,0],y=D[:,1]) # Call the Seaborn plotting directly (requires x= and y= )
#plt.plot(D[:,0],D[:,1],'o') # using matplotlib
#plt.show()
```

```python
df = pd.DataFrame(data=D,columns=['var1','var2']) # Place the D data into a Pandas "data frame" AND provides the "labels" for the x and y axes.

sns.jointplot(x=df.columns[0],y=df.columns[1],data=df, kind='kde') # Call the Seaborn plotting directly (requires x= and y=) other options also exist for jointplot method like 'kind'
plt.show()
```

```python
### Scatterplots ###
x = np.linspace(-1,1,n)
y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2)

sns.scatterplot(x=y1,y=y2,hue=y3,legend=False,palette='bright') # Other palettes: Spectral, viridis, bright, cubehelix

plt.show()
```

```python
### Excercise: Visualize using the Hilbert Matrix in Seaborn using a heatmap...
## H = 1/i+j-1 ##
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

# H = np.zeros((10,10))
# Create a 10x10 Hilbert matrix
H = np.array([[1 / ((i+1) + (j+1) - 1) for j in range(10)] for i in range(10)])

# Visualize using Seaborn
sns.heatmap(H, annot=True, vmax=.40, cmap='Spectral')
plt.show()
```
